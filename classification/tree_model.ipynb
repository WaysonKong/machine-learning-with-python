```python
from sklearn import tree
import graphviz 
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
```

### load data

```python
# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :2]  # we only take the first two features.
y = iris.target
# two class
y_mod_index = [y<=1]
y = y[y_mod_index]
X = X[y_mod_index] 
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
```

```python
### plot first two feature of data set
plt.figure(2, figsize=(8, 6))
plt.clf()

# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1,
            edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xticks(())
plt.yticks(())
plt.show()
```

### Decision tree

#### use first two feature

```python
clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=3)
clf = clf.fit(X, y)
print("score=" + str(clf.score(X,y)))
# 画图
dot_data = tree.export_graphviz(clf, out_file=None)
graph = graphviz.Source(dot_data)  
graph 
```

## Ensemble model

```python
# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :4]  # we only take the first two features.
y = iris.target
# two class
y_mod_index = [y<=5]
y = y[y_mod_index]
X = X[y_mod_index] 
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
```

### Decision Tree

```python
clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=3)
clf = clf.fit(X, y)
dot_data = tree.export_graphviz(clf, out_file=None)
print("score=" + str(clf.score(X,y)))
print("feature importance: ")
print("\t".join([str(x) for x in clf.feature_importances_]))

graph = graphviz.Source(dot_data)  
graph
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth=3, random_state=0, criterion='entropy', n_estimators=3, min_samples_leaf=1)
rf.fit(X, y)
print("score=" + str(rf.score(X,y)))
print("feature importance: ")
print("\t".join([str(x) for x in rf.feature_importances_]))

# figure
dot_data = tree.export_graphviz(rf.estimators_[2], out_file=None)

graph = graphviz.Source(dot_data)  
graph
```

### AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
bst = AdaBoostClassifier(DecisionTreeClassifier(max_depth=3,criterion='entropy'), n_estimators=3)
bst.fit(X, y)
print("score=" + str(bst.score(X,y)))
print("feature importance: ")
print("\t".join([str(x) for x in bst.feature_importances_]))
dot_data = tree.export_graphviz(bst.estimators_[1], out_file=None)

graph = graphviz.Source(dot_data)  
graph
```
