# Linear Model For Regression
### 1. Lasso
$$\min_{w}\frac{1}{N}\|Xw-y\|_2^2 + \alpha \|w\|_1$$
### 2. Ridge Regression
$$\min_{w}\frac{1}{N}\|Xw-y\|_2^2 + \alpha \|w\|_2$$
### 3. [Regression metrics](http://scikit-learn.org/stable/modules/model_evaluation.html#regression-metrics)
* explained_variance_score
* mean_absolute_error
* mean_squared_error
* median_absolute_error
* r2_score

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn import linear_model
```

### Data Preparation: Generate train sample with spark coeffieients
* Data = train set + test set

```python
n_samples, n_features = 50, 200
X = np.random.randn(n_samples, n_features)
coef = 3 * np.random.randn(n_features)
inds = np.arange(n_features)
np.random.shuffle(inds)
coef[inds[10:]] = 0  # sparsify coef
y = np.dot(X, coef)

# add noise
y += 0.01 * np.random.normal(size=n_samples)
```

```python
# Split data in train set and test set
n_samples = X.shape[0]
X_train, y_train = X[:n_samples // 2], y[:n_samples // 2]
X_test, y_test = X[n_samples // 2:], y[n_samples // 2:]
```

### Ridge Regression

```python
reg = linear_model.Ridge(alpha = .5)
y_pred_ridge = reg.fit(X_train, y_train).predict(X_test)
r2_score_ridge = r2_score(y_test, y_pred_ridge)
print(reg)
print("r^2 on test data : %f" % r2_score_ridge)
```

### Lasso

```python
from sklearn.linear_model import Lasso

alpha = 0.1
lasso = Lasso(alpha=alpha)

y_pred_lasso = lasso.fit(X_train, y_train).predict(X_test)
r2_score_lasso = r2_score(y_test, y_pred_lasso)
print(lasso)
print("r^2 on test data : %f" % r2_score_lasso)
```

## Plot coefficients of model

```python
# #############################################################################
plt.plot(lasso.coef_, color='gold', linewidth=2,
         label='Lasso coefficients')
plt.plot(coef, '--', color='navy', label='original coefficients')
plt.plot(reg.coef_, '--', color='red', label='ridge coefficients')
plt.legend(loc='best')
plt.title("Coefficient of Lasso/Ridge Regression")
plt.show()
```
